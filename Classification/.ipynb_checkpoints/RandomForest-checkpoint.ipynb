# Random Forest
This practical example using the RandomForestClassifier from the scikit-learn library to classify whether a person earns more than $50,000 based on certain features.

#### 1. Import the required libraries.
Import necessary libraries for data manipulation, visualization, and linear regression.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

#### 2. Load data from csv to pnadas dataset. 

```python
data = {
    'Age': [25, 32, 47, 51, 62, 23, 31, 40, 34, 58, 26, 29, 36, 45, 54, 33, 38, 44, 56, 60],
    'Experience': [1, 5, 20, 23, 30, 0, 4, 15, 10, 27, 2, 6, 11, 18, 25, 3, 9, 12, 26, 28],
    'Education': [12, 16, 16, 18, 20, 12, 14, 16, 16, 18, 14, 15, 17, 19, 20, 13, 16, 18, 20, 21],
    'Income>50K': [0, 0, 1, 1, 1, 0, 0, 1, 0, 1, 0, 0, 1, 1, 1, 0, 0, 1, 1, 1]  # Binary target variable
}

data = pd.DataFrame(data)
```

#### 3. Visualize Data correlation: 
Plot a correlation matrix to explain relations between variables.

```python
corr_matrix = data.corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

#### 4. Split Data
Split the dataset into training and testing sets.

```python
X = data[['Age', 'Experience', 'Education']]
y = data['Income>50K']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### 5. Train the Linear Regression Model
Create and train a linear regression model using the training data.

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

#### 6. Make Predictions
Use the trained model to predict salaries on the test set.

```python
y_pred = rf_model.predict(X_test)
```

#### 7. Evaluate the Model
Calculate Mean Squared Error and R-squared to evaluate model performance.

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Output the results
print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

- The model achieved perfect classification on this test set, correctly predicting all instances. This might indicate that the test set is small or not challenging enough for the model, or it could be an indication of overfitting if the training and test sets are not appropriately separated.

- If the dataset is small or if the test set is not representative of the problem domain, the perfect score might not generalize to new data. It's crucial to verify the model's performance with cross-validation or a larger, more diverse test set.

- With only six instances in the test set, a perfect accuracy might not be indicative of the model's performance on larger or more complex datasets.

#### 8. Visualize Results
Plot Feature importance

```python
# Feature importance
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]

print("Feature importances:")
for f in range(X.shape[1]):
    print(f"{X.columns[indices[f]]}: {importances[indices[f]]:.4f}")

# Plot feature importances
plt.figure(figsize=(8, 6))
plt.title('Feature Importances')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), X.columns[indices])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.show()
```
