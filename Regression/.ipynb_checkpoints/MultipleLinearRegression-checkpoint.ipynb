# Multiple Linear Regression
This practical example showcasing step-by-step description of multiple linear regression

#### 1. Import the required libraries.
Import necessary libraries for data manipulation, visualization, and linear regression.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

#### 2. Load data from csv to pnadas dataset. 

```python
data = pd.read_csv('Data/salary_data.csv')
data.head(10)
```

#### 3. Visualize Data: 
Plot a scatter plot to visualize the relationship between the variables.

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.scatter(data['Experience Years'], data['Salary'], color='blue')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')

plt.subplot(1, 3, 2)
plt.scatter(data['Education'], data['Salary'], color='green')
plt.xlabel('Education')
plt.ylabel('Salary')
plt.title('Salary vs Education')

plt.subplot(1, 3, 3)
plt.scatter(data['Children'], data['Salary'], color='red')
plt.xlabel('Children')
plt.ylabel('Salary')
plt.title('Salary vs Children')

plt.tight_layout()
plt.show()
```

#### 4. Split Data
Split the dataset into training and testing sets.

```python
y = data['Salary']
X = data[['Experience Years', 'Education', 'Children']]
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

#### 5. Train the Linear Regression Model
Create and train a linear regression model using the training data.

```python
LR_model = LinearRegression()
LR_model.fit(X_train, y_train)
```

#### 6. Make Predictions
Use the trained model to predict salaries on the test set.

```python
y_pred = LR_model.predict(X_test)
```

#### 7. Evaluate the Model
Calculate Mean Squared Error and R-squared to evaluate model performance.

```python
print('MSE:', mean_squared_error(y_test,y_pred))
print('R2:', r2_score(y_test, y_pred))
```

#### 8. Visualize Results
Plot the regression line along with the original data points to visualize the fit.

```python
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, color='blue')
plt.hlines(y=0, xmin=y_pred.min(), xmax=y_pred.max(), linestyles='dashed')
plt.xlabel('Predicted Salary')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Salary')
plt.show()
```
