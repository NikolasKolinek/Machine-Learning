# Logistic Regression
This practical example using Python to predict whether a person earns more than $60,000 per year based on some features using the logistic regression model.

#### 1. Import the required libraries.
Import necessary libraries for data manipulation, visualization, and linear regression.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

#### 2. Load data from csv to pnadas dataset. 

```python
data = pd.read_csv('Data/salary_data.csv')
data.head(10)
```

#### 3. Visualize Data correlation: 
Plot a correlation matrix to explain relations between variables.

```python
corr_matrix = data.corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

We can see Experience Years and Salary have strong (0.98) positive corellation. Also Salary and Education showing correlation (-0.32) and Salary and Income>60K (0.74). Therefore, we choose these two independet variables.

#### 4. Split Data
Split the dataset into training and testing sets.

```python
X = data[['Experience Years', 'Education']]
y = data['Income>60K']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### 5. Train the Linear Regression Model
Create and train a linear regression model using the training data.

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

#### 6. Make Predictions
Use the trained model to predict salaries on the test set.

```python
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]
```

#### 7. Evaluate the Model
Calculate Mean Squared Error and R-squared to evaluate model performance.

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Output the results
print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

- The model tends to predict class 1 more often, resulting in higher precision and recall for class 1 but lower recall for class 0.
- The model has difficulty correctly identifying instances of class 0, as shown by the lower recall and F1-score.
- While the accuracy is 67%, this may not be an ideal measure if classes are imbalanced or if class 0 misclassification is critical.

#### 8. Visualize Results
Plot the regression line along with the original data points to visualize the fit.

```python
# Visualize the predicted probabilities
plt.figure(figsize=(8, 6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Actual')
plt.scatter(range(len(y_pred_proba)), y_pred_proba, color='red', label='Predicted Probability')
plt.title('Actual vs. Predicted Probabilities')
plt.xlabel('Sample Index')
plt.ylabel('Probability of Income > 50K')
plt.legend()
plt.show()
```
