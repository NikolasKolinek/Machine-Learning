# Multiple Linear Regression
This practical example showcasing step-by-step description of multiple linear regression

#### 1. Import the required libraries.
Import necessary libraries for data manipulation, visualization, and linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, root_mean_squared_error
```

#### 2. Load data from csv to pnadas dataset. 

```python
data = pd.read_csv('Data/salary_data.csv')
data.head(10)
```

#### 3. Visualize Data correlation: 
Plot a correlation matrix to explain relations between variables.

```python
corr_matrix = data.corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

We can see Experience Years and Salary have strong (0.98) positive corellation. Also Salary and Education showing correlation (-0.32). Therefore, we choose these two independet variables. 

#### 4. Split Data
Split the dataset into training and testing sets.

```python
y = data['Salary']
X = data[['Experience Years', 'Education']]
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

#### 5. Train the Linear Regression Model
Create and train a linear regression model using the training data.

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

#### 6. Make Predictions
Use the trained model to predict salaries on the test set.

```python
y_pred = model.predict(X_test)
```

#### 7. Evaluate the Model
Calculate Mean Squared Error and R-squared to evaluate model performance.

```python
print('RMSE:', root_mean_squared_error(y_test,y_pred))
print('R2:', r2_score(y_test, y_pred))
```

Approximately 95.68% of the variability in the dependent variable can be explained by the model. With average deviation from the actual salaries by approximately $5,120.64
